```python
x=np.array([3, 4, 5, 6, 8, 9, 10, 11, 12, 14, 15, 16, 17])
y=np.array([1.4, 1.5, 2.2, 2.4, 3.1, 3.2, 3.2, 3.9, 4.1, 4.7, 4.5, 5.2, 5]) 

plt.plot(x, y, 'o')
plt.xlabel('Age')
plt.ylabel('Wing Length')

m, b = np.polyfit(x, y, 1)
plt.plot(x, m*x+b)
```

```python

def f_test(x, y):
    f = np.var(x, ddof=1)/np.var(y, ddof=1) #calculate F test statistic 
    dfn = x.size-1 #define degrees of freedom numerator 
    dfd = y.size-1 #define degrees of freedom denominator 
    p = 1-st.f.cdf(f, dfn, dfd) #find p-value of F test statistic 
    return f, p

print(f_test(x, y))
print('p-value is less than 0.05 so we can reject null')
```
